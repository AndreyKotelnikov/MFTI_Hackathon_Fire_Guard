```python
import os
# import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from ultralytics import YOLO
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
data_yaml_path = "data.yaml"
model_name = "yolov8n.pt"
epochs = 1
image_size = 128

model = YOLO(model_name)
model.train(data=data_yaml_path, epochs=epochs, imgsz=image_size)
```

```python
# –ó–∞–≥—Ä—É–∑–∫–∞ –Ω–∞–∏–ª—É—á—à–µ–π –≤–µ—Ä—Å–∏–∏ –º–æ–¥–µ–ª–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
yolo_model = YOLO('./models/yolo-8epochs-image128px.pt')

test_image_dir = './datasets/test/images'
predicts_for_files = []
y_pred = []
y_test = []

def get_basefilename(filename):
    if filename.lower().endswith('.jpeg'):
        return filename[:-5]
    return filename[:-4]

# –ù—É–∂–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –º–∞—Å—Å–∏–≤ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π Y 
# (–≥–¥–µ –∫–∞–∂–¥–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –≤ –Ω–∞–ª–∏—á–∏–∏ –ø–æ–∂–∞—Ä–∞ –≤ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏)
# –≤ –æ–¥–Ω–æ —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –±—É–¥–µ—Ç –æ—Ç—Ä–∞–∂–∞—Ç—å –æ–±—â—É—é —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –≤ —Ç–æ–º, —á—Ç–æ ‚Äú–≥–¥–µ-—Ç–æ —á—Ç–æ-—Ç–æ –µ—Å—Ç—å‚Äù –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏.
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è —Å–æ–±—ã—Ç–∏–π (–ò–õ–ò)
def calculate_combined_probability(confidences):
    if not confidences:
        # –ï—Å–ª–∏ –Ω–µ—Ç –æ–±—ä–µ–∫—Ç–æ–≤, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 0
        return 0

    # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è –ø–æ–∂–∞—Ä–∞ –≤ –∫–∞–∂–¥–æ–π –æ–±–ª–∞—Å—Ç–∏: 1 - confidence
    no_fire_probabilities = [1 - conf for conf in confidences]

    # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –Ω–µ—Ç –ø–æ–∂–∞—Ä–∞ –Ω–∏–≥–¥–µ:
    combined_no_fire_prob = np.prod(no_fire_probabilities)

    # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Ö–æ—Ç—è –±—ã –≥–¥–µ-—Ç–æ –µ—Å—Ç—å –ø–æ–∂–∞—Ä:
    combined_probability = 1 - combined_no_fire_prob
    return combined_probability

# i = 0
for filename in os.listdir(test_image_dir):
    if filename.endswith(('.jpg', '.jpeg', '.png')):  # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º —Ç–æ–ª—å–∫–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
        image_path = os.path.join(test_image_dir, filename)
        try:
            results = yolo_model.predict(source=image_path)
            confidences = []
            for result in results:
                boxes = result.boxes
                if boxes:
                    confidences.extend([box.conf.item() for box in boxes])
            predicts_for_files.append({
                'basefilename': get_basefilename(filename),
                'pred': calculate_combined_probability(confidences),
                'confidences': confidences,
                'test': None
            })
        except Exception as e:
            print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è {filename}: {e}")
        # i += 1
        # if i >= 200:
        #     break

test_labels_dir = './datasets/test/labels'
for predicts in predicts_for_files:
    filename = predicts['basefilename'] + '.txt'
    try:
        with open(os.path.join(test_labels_dir, filename), "r") as f:
            lines = f.readlines()
            if lines:
                data = lines[0].strip().split()
                confidence = 1 if len(data) > 0 else 0
                y_test.append(confidence)
                predicts['test'] = confidence
            else:
                predicts['test'] = 0
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —á—Ç–µ–Ω–∏–∏ —Ñ–∞–π–ª–∞ {filename}: {e}")

# ----------------------------------------------------------------------------------
# –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –º–æ–¥–µ–ª–∏ —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ
# –µ—Å–ª–∏ –æ–Ω–∞ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –∫–∞–∫—É—é-—Ç–æ –æ–±–ª–∞—Å—Ç—å, –∑–Ω–∞—á–∏—Ç –æ–Ω–∞ —É–∂–µ —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç —á—Ç–æ —á—Ç–æ-—Ç–æ –Ω–∞—à–ª–∞, —Ç.–µ. pred > 0.5
# –≠—Ç–æ –≤–∞–∂–Ω–æ, –ø–æ—Ç–æ–º—É —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å –∫–æ—Ç–æ—Ä–æ–π –º–æ–¥–µ–ª—å –¥–∞—ë—Ç –æ—Ç–≤–µ—Ç –≤–∫–ª—é—á–∞–µ—Ç –≤ —Å–µ–±—è –Ω–µ —Ç–æ–ª—å–∫–æ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –≤ –Ω–∞–ª–∏—á–∏–∏ –ø–æ–∂–∞—Ä–∞
# –Ω–æ –∏ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –≤ –º–µ—Å—Ç–µ –µ–≥–æ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏—è.
# –¢–∞–∫–∞—è "—É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å" –º–æ–¥–µ–ª–∏ –≤—Å–µ–≥–¥–∞ –±—É–¥–µ—Ç –∑–∞–≤–µ–¥–æ–º–æ –º–µ–Ω—å—à–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø—Ä–æ—Å—Ç–æ –Ω–∞–ª–∏—á–∏—è –ø–æ–∂–∞—Ä–∞
for x in predicts_for_files:
    x['pred0'] = (1 + x['pred']) / 2 if x['pred'] > 0 else 0
# ----------------------------------------------------------------------------------

y_pred = [1 if x['pred0'] >= 0.5 else 0 for x in predicts_for_files]
y_test = [x['test'] for x in predicts_for_files]
y_pred, y_test = np.array(y_pred), np.array(y_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='binary')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")

report = classification_report(y_test, y_pred, target_names=["Class 0", "Class 1"])

print("\nClassification Report:\n")
print(report)
```
